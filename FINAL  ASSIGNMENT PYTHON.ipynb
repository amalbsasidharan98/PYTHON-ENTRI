```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the Excel file
file_path = "myexcel.xlsx"
excel_file = pd.ExcelFile(file_path)

print(excel_file.sheet_names)

df = pd.read_excel(file_path, sheet_name='myexcel.csv')  # or use index: sheet_name=0

np.random.seed(42)
df['Height'] = np.random.randint(150, 181, size=len(df))
```

```python
#TASK 2 Segregate Employees by Position

position_counts = df['Position'].value_counts().reset_index()
position_counts.columns = ['Position', 'Count']

plt.figure(figsize=(8, 5))
sns.barplot(data=position_counts, x='Position', y='Count', palette='plasma')
plt.title('Employee Count by Position')
plt.xlabel('Position')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
#3 Predominant Age Group

#Define age groups
bins = [18, 25, 30, 35, 40, 45, 50]
labels = ['19-25', '26-30', '31-35', '36-40', '41-45', '46-50']
df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=True)

# Count by age group
age_group_counts = df['Age Group'].value_counts().sort_index().reset_index()
age_group_counts.columns = ['Age Group', 'Count']

# Display the data
display(age_group_counts)

# Plot
plt.figure(figsize=(8, 5))
sns.barplot(data=age_group_counts, x='Age Group', y='Count', palette='coolwarm')
plt.title('Employee Count by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
#4 Salary Expenditure by Team and Position

# Handle any missing salaries
df['Salary'] = pd.to_numeric(df['Salary'], errors='coerce')  
df['Salary'].fillna(0, inplace=True)

# Group by Team
team_salary = df.groupby('Team')['Salary'].sum().sort_values(ascending=False).reset_index()

# Group by Position
position_salary = df.groupby('Position')['Salary'].sum().sort_values(ascending=False).reset_index()

# Display top team and position
print("🔝 Team with highest salary expenditure:")
display(team_salary.head(1))

print("🔝 Position with highest salary expenditure:")
display(position_salary.head(1))
```

```python

 Plot - Team Salary
plt.figure(figsize=(12, 6))
sns.barplot(data=team_salary, x='Team', y='Salary', palette='magma')
plt.title('Total Salary Expenditure by Team')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# Plot - Position Salary
plt.figure(figsize=(8, 5))
sns.barplot(data=position_salary, x='Position', y='Salary', palette='viridis')
plt.title('Total Salary Expenditure by Position')
plt.tight_layout()
plt.show()
```

```python
#5 Correlation Between Age and Salary

# Ensure Salary column is clean
df['Salary'] = pd.to_numeric(df['Salary'], errors='coerce')
df['Salary'].fillna(0, inplace=True)

# Calculate correlation
correlation = df['Age'].corr(df['Salary'])
print(f"🔗 Correlation between Age and Salary: {correlation:.2f}")
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='Age', y='Salary', alpha=0.6)
sns.regplot(data=df, x='Age', y='Salary', scatter=False, color='red')  # trend line
plt.title('Correlation Between Age and Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.tight_layout()
plt.show()
```
